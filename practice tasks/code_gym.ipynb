```python

import pandas as pd

data = {
    "a": [3, 8, 6, 2],
    "b": [5, 4, 7, 9],
    "c": ["yes", "no", "yes", "no"],
    "d": [2, 7, 4, 6],
    "e": ["low", "high", "low", "medium"],
    "f": [100, 200, 150, 180]
}

df = pd.DataFrame(data)
df
```

```python
df["y"] = 0

df
```

```python
def compute_target(row):
    if (row["a"] + row["b"] > 10) and (row["c"] == "yes"):
        return "fault 2"
    elif (row["d"] > 5) or (row["e"] == "high"):
        return "fault 1"
    else:
        return "fault 3"
```

```python
df["y"] = df.apply(compute_target, axis=1)

df
```
