```python
from langchain_community.document_loaders import PyPDFLoader

path = "../assignments/Ayyub_Raji.pdf"

def load_file(path):
    
    loader = PyPDFLoader(path)

    pages = loader.load()

    return pages
```

```python
document_pages = load_file(path)

document_pages
```

```python
def print_document(pages_list):
    if pages_list:
        for page in pages_list:
            document = page.page_content
            print(document)
            
    
```

```python
print_document(document_pages)
```

**Chunking** with rcts

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)

chunked_doc = splitter.split_documents(document_pages)
chunked_doc
```
